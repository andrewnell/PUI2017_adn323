```python
# Note: I am adding imports here, to show the workflow: 
# at this point I realize what packages I need and I import them. 
# but this is a PEP8 violation. 
# It is a good habit to put the includes at the top 
# so the code fails right away if the packages are missing. 
# Otherwise a user may have a piece of code running for a long time, 
# and then fail when the package cannot be imported... 
# (unless they are only imported under conditions, e.g. in an if statement)

import numpy as np
Distributions = {}
np.random.seed(100) # Create random seed for repeatability
# Distributions is an empty disctionary
Distributions['chisq'] = np.random.chisquare(df, size=100)
# Distributions now is a dictionary with one object: an array by the key 'chisq'
pl.hist(Distributions['chisq'], bins = 30)
pl.ylabel('N')
pl.xlabel('x')
```

## Now let's do it for 100 Chisq distributions
we want distributions of different sizes (all integers between 10 and 2000)

```python
np.random.seed(0)
# mysize = (2000 / (np.array(range(1, 100)))).astype(int)
mysize = np.random.randint(10,2000,100)
# mysize = 1000 / (np.array(range(1, 100) + [10]))
# mysize = (np.random.rand(100) * 1000).astype(int)
print (mysize, mysize.shape)
```

```python
# I am gonna overwrite the disctionary entry created into a nested dictionary 
# the outside dictionary hosts the distributions as dictionary, 
# and each distribution dictionary contains the 100 distribution (arrays of diferent sizes)

Distributions['chisq'] = {} 

np.random.seed(0) # Create random seed for repeatability

#and do it in a for loop. not pythonic, but easily readable

for n in mysize:
    Distributions['chisq'][n] = np.random.chisquare(df, size = n)
    
# save the means for plotting later    

Distributions['chisq']['means'] = {}
axchisq_mu_n = pl.figure(figsize=(10,6)).add_subplot(111)

for nn in Distributions['chisq']:
    if not type(nn) == str:
        Distributions['chisq']['means'][nn] = Distributions['chisq'][nn].mean()
        #and plot it
        axchisq_mu_n.plot(nn, Distributions['chisq']['means'][nn], 'o')
        axchisq_mu_n.set_xlabel('sample size', fontsize=18)
        axchisq_mu_n.set_ylabel('sample mean', fontsize=18)
        axchisq_mu_n.set_title('Chi squared', fontsize=18)
        axchisq_mu_n.plot([min(mysize), max(mysize)], [mymean, mymean], 'k')
```

### Figure 2: Random Chi Squared distributions and with varying sample sizes between 10 and 2000. 
#### Note how as sample size increases so does the accuracy of the mean as stated by the law of large numbers

# now the histogram of the mean of all samples: 
## this requires some data wrangling, since i put the data in a dictionary which is a user friendly data structure, but perhaps not the most machine friendly one

a coder friendly, though not terribly pythonic way to do it is to append each mean in a list in a for loop, and then hist that. be careful: if your data size gets large this becomes a very bad idea!!

```python
chisqmeans = list(Distributions['chisq']['means'].values())

###############################################################################
# create gaussian curve fitting variables in dictionary for later use

Distributions['chisq']['gaussianmean'] = {}
Distributions['chisq']['gaussianstd'] = {}

#create gaussian variables 
Distributions['chisq']['gaussianmean'] = np.mean(chisqmeans)
Distributions['chisq']['gaussianstd'] = np.std(chisqmeans)

###############################################################################
# Plot graph with distribution of means
pl.figure(figsize=(10, 10))
pl.hist(chisqmeans,bins=30,normed=True)
pl.title('Chi Squared Mean Distributions', fontsize = 18)
pl.xlabel('sample mean', fontsize = 18)
pl.ylabel('N', fontsize = 18)

###############################################################################
#Prepare gaussian fitting plot and plot
xmin, xmax = plt.xlim()
xlim = np.linspace(xmin, xmax, 100)
fit = nm.pdf(xlim,Distributions['chisq']['gaussianmean'] ,
             Distributions['chisq']['gaussianstd'])
pl.plot(xlim, fit,'--b')

# assisted with code from https://stackoverflow.com/questions/20011122/
# fitting-a-normal-distribution-to-1d-data 
# and assistance from Ruben Hambardzumyan
```

# Do it for the Normal distribution


```python
###############################################################################
np.random.seed(0) # Create random seed for repeatability

Distributions['normal'] = {} # Create empty dictionary position for normal

# Create Normal distributions for each sample size
for n in mysize:
    Distributions['normal'][n] = np.random.normal(mymean, np.random.random(),n)
    
###############################################################################
    
Distributions['normal']['means'] = {} # Create empty means slot in dictionary

# plot means of all distributions
axnorm_mu_n = pl.figure(figsize=(10,6)).add_subplot(111)

for nn in Distributions['normal']:
    if not type(nn) == str:
        Distributions['normal']['means'][nn] = (Distributions['normal']
                                                [nn].mean())
        #and plot it
        axnorm_mu_n.plot(nn, Distributions['normal']['means'][nn], 'o')
        axnorm_mu_n.set_xlabel('sample size', fontsize=18)
        axnorm_mu_n.set_ylabel('sample mean', fontsize=18)
        axnorm_mu_n.set_title('Normal', fontsize=18)
        axnorm_mu_n.plot([min(mysize), max(mysize)], [mymean, mymean], 'k')
```

### Figure 4: Random Normal distributions and with varying sample sizes between 10 and 2000. 
#### Note how as sample size increases so does the accuracy of the mean as stated by the law of large numbers

```python
normmeans = list(Distributions['normal']['means'].values())

###############################################################################
# create gaussian curve fitting variables in dictionary for later use

Distributions['normal']['gaussianmean'] = {}
Distributions['normal']['gaussianstd'] = {}

#create gaussian variables 
Distributions['normal']['gaussianmean'] = np.mean(normmeans)
Distributions['normal']['gaussianstd'] = np.std(normmeans)

###############################################################################
# Plot graph with distribution of means
pl.figure(figsize=(10, 10))
pl.hist(normmeans,bins=30,normed=True)
pl.title('Normal Mean Distributions', fontsize = 18)
pl.xlabel('sample mean', fontsize = 18)
pl.ylabel('N', fontsize = 18)

###############################################################################
#Prepare gaussian fitting plot and plot
xmin, xmax = plt.xlim()
xlim = np.linspace(xmin, xmax, 100)
fit = nm.pdf(xlim,Distributions['normal']['gaussianmean'] ,
             Distributions['normal']['gaussianstd'])
pl.plot(xlim, fit,'--b')

# assisted with code from https://stackoverflow.com/questions/20011122/
# fitting-a-normal-distribution-to-1d-data 
# and assistance from Ruben Hambardzumyan
```

# Do it for the Poisson distribution

```python
###############################################################################
np.random.seed(0) # Create random seed for repeatability

Distributions['pois'] = {} # Create empty dictionary position for poisson

# Create Poisson distributions for each sample size
for n in mysize:
    Distributions['pois'][n] = np.random.poisson(mymean, n)
    
###############################################################################
    
Distributions['pois']['means'] = {} # Create empty means slot in dictionary

# plot means of all distributions
axpois_mu_n = pl.figure(figsize=(10,6)).add_subplot(111)

for nn in Distributions['pois']:
    if not type(nn) == str:
        Distributions['pois']['means'][nn] = (Distributions['pois'][nn].mean())
        #and plot it
        axpois_mu_n.plot(nn, Distributions['pois']['means'][nn], 'o')
        axpois_mu_n.set_xlabel('sample size', fontsize=18)
        axpois_mu_n.set_ylabel('sample mean', fontsize=18)
        axpois_mu_n.set_title('Poisson', fontsize=18)
        axpois_mu_n.plot([min(mysize), max(mysize)], [mymean, mymean], 'k')
```

### Figure 6: Random Poisson distributions and with varying sample sizes between 10 and 2000. 
#### Note how as sample size increases so does the accuracy of the mean as stated by the law of large numbers

```python
poismeans = list(Distributions['pois']['means'].values())

###############################################################################
# create gaussian curve fitting variables in dictionary for later use

Distributions['pois']['gaussianmean'] = {}
Distributions['pois']['gaussianstd'] = {}

#create gaussian variables 
Distributions['pois']['gaussianmean'] = np.mean(poismeans)
Distributions['pois']['gaussianstd'] = np.std(poismeans)

###############################################################################
# Plot graph with distribution of means
pl.figure(figsize=(10, 10))
pl.hist(poismeans,bins=30,normed=True)
pl.title('Poisson Mean Distributions', fontsize = 18)
pl.xlabel('sample mean', fontsize = 18)
pl.ylabel('N', fontsize = 18)

###############################################################################
#Prepare gaussian fitting plot and plot
xmin, xmax = plt.xlim()
xlim = np.linspace(xmin, xmax, 100)
fit = nm.pdf(xlim,Distributions['pois']['gaussianmean'] ,
             Distributions['pois']['gaussianstd'])
pl.plot(xlim, fit,'--b')

# assisted with code from https://stackoverflow.com/questions/20011122/
# fitting-a-normal-distribution-to-1d-data 
# and assistance from Ruben Hambardzumyan
```

# Do it for the Binomial distribution

```python
###############################################################################
np.random.seed(0) # Create random seed for repeatability

Distributions['bin'] = {} # Create empty dictionary position for binomial

# Create Binomial distributions for each sample size
trials = np.random.randint(100,100000)

for n in mysize:
    Distributions['bin'][n] = np.random.binomial(trials, mymean/float(trials), n)
    
###############################################################################
    
Distributions['bin']['means'] = {} # Create empty means slot in dictionary

# plot means of all distributions
axbin_mu_n = pl.figure(figsize=(10,6)).add_subplot(111)

for nn in Distributions['bin']:
    if not type(nn) == str:
        Distributions['bin']['means'][nn] = (Distributions['bin'][nn].mean())
        #and plot it
        axbin_mu_n.plot(nn, Distributions['bin']['means'][nn], 'o')
        axbin_mu_n.set_xlabel('sample size', fontsize=18)
        axbin_mu_n.set_ylabel('sample mean', fontsize=18)
        axbin_mu_n.set_title('Binomial', fontsize=18)
        axbin_mu_n.plot([min(mysize), max(mysize)], [mymean, mymean], 'k')
```

### Figure 8: Random Binomial distributions and with varying sample sizes between 10 and 2000. 
#### Note how as sample size increases so does the accuracy of the mean as stated by the law of large numbers

```python
binmeans = list(Distributions['bin']['means'].values())

###############################################################################
# create gaussian curve fitting variables in dictionary for later use

Distributions['bin']['gaussianmean'] = {}
Distributions['bin']['gaussianstd'] = {}

#create gaussian variables 
Distributions['bin']['gaussianmean'] = np.mean(binmeans)
Distributions['bin']['gaussianstd'] = np.std(binmeans)

###############################################################################
# Plot graph with distribution of means
pl.figure(figsize=(10, 10))
pl.hist(binmeans,bins=30,normed=True)
pl.title('Binomial Mean Distributions', fontsize = 18)
pl.xlabel('sample mean', fontsize = 18)
pl.ylabel('N', fontsize = 18)

###############################################################################
#Prepare gaussian fitting plot and plot
xmin, xmax = plt.xlim()
xlim = np.linspace(xmin, xmax, 100)
fit = nm.pdf(xlim,Distributions['bin']['gaussianmean'] ,
             Distributions['bin']['gaussianstd'])
pl.plot(xlim, fit,'--b')

# assisted with code from https://stackoverflow.com/questions/20011122/
# fitting-a-normal-distribution-to-1d-data 
# and assistance from Ruben Hambardzumyan
```
